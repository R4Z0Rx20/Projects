## predicting tip from total bill using simple linear regression

```python
r = np.corrcoef(tips['total_bill'], tips['tip'])[0, 1]
plt.scatter(tips['total_bill'], tips['tip'], label = f'r : {r}')
plt.legend()
plt.show()
```

$\hat { \text {tip} } = \theta_0 + \theta_1 *  \text {total_bill}$

$\theta = {(X^TX)}^{-1}X^Ty$ vector format

$$\theta_1 =  \frac { \sum (\bar y - y^{(i)}) (\bar X_1 - {X_1}^{(i)}) } { \sum {(\bar X_1 - {X_1}^{(i)})}^2 } $$

$$\theta_0 =\bar y -  \theta_1 \bar X_1 $$

```python
X1 = np.array(tips['total_bill'])
y = np.array(tips['tip'])
```

```python
theta_1 =  ( (y.mean() - y)*(X1.mean() - X1) ).sum() / ( ( X1.mean() - X1)**2 ).sum()
theta_0 =  y.mean() - theta_1 * X1.mean()
print(f'theta_1 : {theta_1}')
print(f'theta_0 : {theta_0}')
y_pred = theta_0 + theta_1 * X1
```

```python
plt.scatter(tips['total_bill'], tips['tip'], label = f'r : {r}')
plt.plot(X1, y_pred, "r-", label="prediction line")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.legend()
plt.show()
```
